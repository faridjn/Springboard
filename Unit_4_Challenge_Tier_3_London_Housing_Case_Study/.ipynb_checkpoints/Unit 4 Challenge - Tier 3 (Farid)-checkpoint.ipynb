```python
#Pass the date values from column 2 to then into a list
date_time_columns = df.columns[2:]

#Change the date formatting by dropping time values
date_columns = [dt.datetime.strftime(x, "%Y-%m-%d") for x in date_time_columns]

#Create a new list with "borough", "id" and extend it by the date_columns list
new_columns = ["london_borough", "id"] + date_columns
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
df_melt = df.melt(id_vars=['london_borough','id'], var_name = ['month'], value_name = "price")
df_melt.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Change the data type of 'price' & 'year'
df_melt['price'] = df_melt['price'].astype(float)
df_melt['month'] = df_melt['month'].astype('datetime64[ns]')
df_melt.info()  #Check data types
```

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column?

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Create a pivot table
df_annual_average = df_final.copy()
df_annual_average['year'] = pd.DatetimeIndex(df_final['month']).year
pivot_table = pd.pivot_table(df_annual_average, values = 'price', index = ['london_borough', 'year'], aggfunc=np.mean)
pivot_table = pivot_table.reset_index()
print(pivot_table.head())
```

```python
#Custom functions

#Faction to calcuate the change ratio
def cal_ratio(df: pd.DataFrame, index_col: str, value_col: str, start_idx: int, end_idx: int):
    ''' This function gets:
        - dataframe
        - index & value column names
        - star index
        - end index
        - return the ration of change for '''
    start_val = float(df[value_col].loc[df[index_col] == start_idx])
    end_val = float(df[value_col].loc[df[index_col] == end_idx])
    ratio_percent = 100*end_val/start_val
    
    return ratio_percent

# Function to calculate max val and it key
def dict_max_item(dct: dict):
    '''This function gets a dictionary and returns the max val and its key'''
    max_val = max(dct.values())
    for k, v in ratio_dct.items():
        if v == max_val:
            max_key = k 
            
    return max_key, max_val
```

```python
#Get the all London borough names
london_borough_names = list(pivot_table['london_borough'].unique())

#Define an empty dictionary
ratio_dct = dict()

start_year = 1998
end_year = 2018

#Iterate the london_borough_names and run the change ratio function
for borough in london_borough_names:
    #Filter the dataframe to the borough
    df_filtered = pivot_table.loc[pivot_table['london_borough'] == borough]
    
    #Calculate change ratio for the filtered dataframe
    val = cal_ratio(df_filtered, 'year', 'price', start_year, end_year)
    
    #Pass the values to dictionary
    ratio_dct[borough] = val
    
#Unpack key and the value for max item
max_key, max_val = dict_max_item(ratio_dct)
```

```python
#import ticker for using % on y lable
import matplotlib.ticker as mtick

#Create and empty list of colours
colors = []

#Set color for london_borough_names
#Set "red" if value is max, otherwise "blue"
for index, value in ratio_dct.items():
    if value == max_val:
        colors.append('#f77088') #red
    else:
        colors.append('#66aabb') #blue
        
fig = plt.figure(figsize = (15, 5))
#Show y unit in %
ax = fig.add_subplot(1,1,1)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

plt.bar(ratio_dct.keys(),ratio_dct.values(), width = .8, color=colors)
plt.xticks(rotation=90)
plt.size=(150, 50)



#Set title labels and title
title = 'London Borough Husing Price Change (' + str(start_year) + ' - ' + str(end_year) +')'
plt.title(title)
plt.xlabel('London Borough')
plt.ylabel('Price Change (%)')


plt.show()
```
