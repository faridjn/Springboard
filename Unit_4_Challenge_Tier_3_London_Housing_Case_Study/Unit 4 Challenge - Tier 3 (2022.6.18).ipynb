**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
#Function to calcuate the change ratio
def cal_ratio(df: pd.DataFrame, index_col: str, value_col: str,
              start_idx: int, end_idx: int):
    ''' This function gets:
        - dataframe
        - index & value column names
        - start index
        - end index
        - return the ration of change for '''
    start_val = float(df[value_col].loc[df[index_col] == start_idx])
    end_val = float(df[value_col].loc[df[index_col] == end_idx])
    ratio_percent = 100 * end_val/start_val
    
    return ratio_percent

# Function to calculate max val and it key
def dict_max_item(dct: dict):
    '''This function gets a dictionary and
    returns the max val and its key'''
    max_val = max(dct.values())
    for k, v in ratio_dct.items():
        if v == max_val:
            max_key = k 
            
    return max_key, max_val
```

```python
#Calculate the change ration for all boroughs between 1998 and 2018

#time frame
start_year = 1998
end_year = 2018

#Define an empty dictionary to pass the ratio to boroughs as {'borough': 'ration'}
ratio_dct = dict()

#Get the all unique London borough names 
london_borough_names = list(dfg['london_borough'].unique())


#Iterate the london_borough_names and run the change ratio function then store in the dictionary
for borough in london_borough_names:
    #Filter the dataframe to the borough
    df_filtered = dfg.loc[dfg['london_borough'] == borough]
    
    #Calculate change ratio for the filtered dataframe
    val = cal_ratio(df_filtered, 'year', 'price', start_year, end_year)
    
    #Pass the values to dictionary
    ratio_dct[borough] = val
```

```python
#Sort the dictionary for a better visuzalization 
ratio_dct_sorted = dict(sorted(ratio_dct.items(), key=lambda item: item[1]))
```

```python
#Find the maximum change ratio and the associated borough
#Unpack key and the value for max item
max_key, max_val = dict_max_item(ratio_dct_sorted)
```

```python
#Plot the ratio changes for borough

#Import ticker for using % on y lable
import matplotlib.ticker as mtick

#The maximum value will be show with a different color.
#So, the plot colors will be dynamically set based on the data

#Create and empty list of colors
colors = []

#Set color for ratio values
#Set "red" if value is max, otherwise "blue" (as default)
for index, value in ratio_dct_sorted.items():
    if value == max_val:
        colors.append('#f77088') #red for max
    else:
        colors.append('#66aabb') #blue as default

#set chart size
fig = plt.figure(figsize = (15, 5))

#Settings to show Y axis unit in %
ax = fig.add_subplot(1,1,1)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

#Pass the x and height values with color set to bar plotter
plt.bar(ratio_dct_sorted.keys(),ratio_dct_sorted.values(), width = .8, color=colors)

#Rotate the x labels for better visuzlaizarion
plt.xticks(rotation=90)


#Set title labels and title
title = 'London Borough Husing Price Change (' + str(start_year) + ' - ' + str(end_year) +')'
plt.title(title)
plt.xlabel('London Borough')
plt.ylabel('Price Change (%)')

#Show the plot
plt.show()
```
